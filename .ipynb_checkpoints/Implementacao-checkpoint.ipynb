```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from skimage.feature import local_binary_pattern
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import itertools
from sklearn.neighbors import DistanceMetric
from sklearn.svm import SVC
import pickle
```

# 1. Loading DataSet

```python
#Dataset Folder
dataset_folder = "dataset/"


names = []
images = []

#Reading all data 
for folder in os.listdir(dataset_folder):
    for name in os.listdir(os.path.join(dataset_folder, folder))[:70]: # limit only 70 face per class
        if name.find(".jpg") > -1 :
            img = cv2.imread(os.path.join(dataset_folder + folder, name))
            #Saving image
            images.append(img)
            #Name of current data (label) 
            names.append(folder)

            
#Labels from names of all data
labels = np.unique(names)
labels
```

# 2. Printing DataSet

```python
def show_dataset(images_class, label):
    # show data for 1 class
    plt.figure(figsize=(14,5))
    k = 0
    for i in range(1,6):
        plt.subplot(1,5,i)
        try :
            plt.imshow(images_class[k][:,:,::-1])
        except :
            plt.imshow(images_class[k], cmap='gray')
        plt.title(label)
        plt.axis('off')
        plt.tight_layout()
        k += 1
    plt.show()
```

```python
for label in labels:

    ids = np.where(label== np.array(names))[0]
    images_class = images[ids[0] : ids[-1] + 1]
    show_dataset(images_class, label)
```

# 3.  Handling Dataset 
### (Crop, Risize and Convert to Gray Scale)

```python
#Face detector algorithm to crop
face_cascade = cv2.CascadeClassifier('haarcascades/haarcascade_frontalface_default.xml')

#Fuction for detect face, change color and resize of some given image
def detect_face(img, idx):
    #Convert to Grayscale
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    #Try to detect face
    faces = face_cascade.detectMultiScale(img, 1.3, 5)
    
    #Verify if a face has been found
    try :
        x, y, w, h = faces[0]
        #Crop image
        img = img[y:y+h, x:x+w]
        #Resize image
        img = cv2.resize(img, (100, 100))
    except :
        print("Face not found in image index", i)
        img = None
    #Return image
    return img
```

```python
croped_images = []
for i, img in enumerate(images) :
    img = detect_face(img, i)
    if img is not None :
        croped_images.append(img)
    else :
        del names[i]
```

```python
for label in labels:
    
    ids = np.where(label== np.array(names))[0]
    images_class = croped_images[ids[0] : ids[-1] + 1] # select croped images for each class
    show_dataset(images_class, label)
```

# 4. Labeling and Splitting Data with Scikit-Learn

```python
label_encoder = LabelEncoder()
label_encoder.fit(names)

print("Labels: ")
for i, label in enumerate(label_encoder.classes_):
    print ("\t", i, label)   

names_labeled = label_encoder.transform(names)
print("Label Code of each image: ")
print( names_labeled)
```

```python
x_train, x_test, y_train, y_test = train_test_split(np.array(croped_images, dtype=np.float32),   # input data
                                                    np.array(names_labeled),                            # target/output data 
                                                    test_size=0.25, 
                                                    random_state=42)
print("Train Images:", x_train.shape)
print("Train Labels:", y_train.shape)
print("Test Images:", x_test.shape)
print("Test Labels:", y_test.shape)
```

```python
class LBPH_SVM_Recognizer_V1():
    #
    # ----- class for LBPH Scikit + SVM OpenCV------
    #
    def __init__(self, max_iter=100, epsilon=0.001, C=100, Gamma=0.001):
        self.svm = cv2.ml.SVM_create()
        self.svm.setKernel(cv2.ml.SVM_CHI2)
        self.svm.setType(cv2.ml.SVM_C_SVC)
        self.svm.setTermCriteria((cv2.TERM_CRITERIA_MAX_ITER, max_iter, epsilon))
        self.svm.setC(C)           
        self.svm.setGamma(Gamma) 
        
        self.face_histograms = []
        self.y = []
    
    def find_lbp_histogram(self, image, P=8, R=1, eps=1e-7, n_window=(8,8)):
        E = []
        h, w = image.shape
        h_sz = int(np.floor(h/n_window[0]))
        w_sz = int(np.floor(w/n_window[1]))
        lbp_img = local_binary_pattern(image, P=P, R=R, method="default")
        for (x, y, C) in self.sliding_window(lbp_img, stride=(h_sz, w_sz), window=(h_sz, w_sz)):
            if C.shape[0] != h_sz or C.shape[1] != w_sz:
                continue
            H = np.histogram(C,                          
                             bins=2**P, 
                             range=(0, 2**P),
                             density=True)[0] 
            E.extend(H)
        return E
    
    def sliding_window(self, image, stride, window):
        for y in range(0, image.shape[0], stride[0]):
            for x in range(0, image.shape[1], stride[1]):
                yield (x, y, image[y:y + window[1], x:x + window[0]])
                   
    def train(self, x, y):
        self.y = y
        self.face_histograms = [self.find_lbp_histogram(img) for img in x]
        hist_mat = np.array(self.face_histograms, dtype=np.float32)
        self.svm.train(hist_mat, cv2.ml.ROW_SAMPLE, y)
    
    def predict(self, x):
        hists = [self.find_lbp_histogram(img) for img in x]
        hist_mat = np.array(hists, dtype=np.float32)
        ret, idx = self.svm.predict(hist_mat, True)
        print("Ret:", ret)
        confidence = 1.0 / (1.0 + np.exp(-ret)) # convert retVal to confidence level (0-1) sigmoid
        return idx, confidence
```

```python
class LBPH_SVM_Recognizer_V2():
    #
    # ----- class for LBPH Scikit + SVM Scikit------
    #
    def __init__(self, C=100, Gamma=0.001):
        self.svm = SVC(kernel='precomputed', C=C, gamma=Gamma, probability=True)
        self.chi2 = DistanceMetric.get_metric('pyfunc', func=self.chi2_distance)
        self.face_histograms = []
        self.hist_mat = []
        
    def chi2_distance(self, hist1, hist2, gamma=0.5): 
        chi = - gamma * np.sum(((hist1 - hist2) ** 2) / (hist1 + hist2 + 1e-7)) 
        return chi

    def find_lbp_histogram(self, image, P=8, R=1, eps=1e-7, n_window=(8,8)):
        E = []
        h, w = image.shape
        h_sz = int(np.floor(h/n_window[0]))
        w_sz = int(np.floor(w/n_window[1]))
        lbp_img = local_binary_pattern(image, P=P, R=R, method="default")
        for (x, y, C) in self.sliding_window(lbp_img, stride=(h_sz, w_sz), window=(h_sz, w_sz)):
            if C.shape[0] != h_sz or C.shape[1] != w_sz:
                continue
            H = np.histogram(C,                          
                             bins=2**P, 
                             range=(0, 2**P),
                             density=True)[0] 
            
            H = H.astype("float")
            H /= (H.sum() + eps)
            E.extend(H)
        return E
    
    def sliding_window(self, image, stride, window):
        for y in range(0, image.shape[0], stride[0]):
            for x in range(0, image.shape[1], stride[1]):
                yield (x, y, image[y:y + window[1], x:x + window[0]])
                   
    def train(self, x, y):
        self.face_histograms = [self.find_lbp_histogram(img) for img in x]
        self.hist_mat = np.array(self.face_histograms, dtype=np.float32)
        K = self.chi2.pairwise(self.hist_mat,self.hist_mat)
        self.svm.fit(K, y)
    
    def predict(self, x):
        hists = [self.find_lbp_histogram(img) for img in x]
        hist_mat = np.array(hists, dtype=np.float32)
        K = self.chi2.pairwise(hist_mat, self.hist_mat)
        idx = self.svm.predict(K)
        
        
        raw_probs = self.svm.predict_proba(K)
        probs = []
        for i in range(len(raw_probs)):
            probs.append(raw_probs[i][idx[i]]*100)
            

        return idx, probs
```

```python
lbph_svm_model_v1 = LBPH_SVM_Recognizer_V2()
lbph_svm_model_v1.train(x_train, y_train)
```

# Util - Capture WebCam Images for DataSet

```python
def detect_all_faces(frame):

    #Convert to Grayscale
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    #Try to detect face
    faces = face_cascade.detectMultiScale(frame, 1.3, 5)
    
    cropped_faces = []
    for (x,y,w,h) in faces:
        img = frame[y:y+h, x:x+w]
        img = cv2.resize(img, (100, 100))
        cropped_faces.append(img)
    
    return faces, cropped_faces
    
```

```python
cap = cv2.VideoCapture("Video_2.mp4")

faces_history = []

while cap.isOpened():
    ret, frame = cap.read()
        
    if ret:
        frame = cv2.resize(frame, (1000,800))
            
            
        #Detect faces
        faces, cropped_faces = detect_all_faces(frame)
    
    
        #Classifies faces
        current_labels = []
        if cropped_faces:
            faces_history.append(cropped_faces)
            faces_predict, probs = lbph_svm_model_v1.predict(cropped_faces)
            for i in range(len(faces_predict)):
            
                #current_labels.append(labels[faces_predict[i]])
                current_labels.append(f"{ labels[faces_predict[i]] }({probs[i]})" )
            
    
        #Draw Faces 
        i = 0
        for (x,y,w,h) in faces:
            frame = cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
            if(probs[i] > 75):
                cv2.putText(frame, current_labels[i], (x,y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255,0,0),2 )
            else:
                cv2.putText(frame, "Unknown", (x,y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255,0,0),2 )
                
            i += 1
            
            
        cv2.imshow("Capture Photo", frame)
        if cv2.waitKey(10) == ord('q'):
            break
         

cap.release()
cv2.destroyAllWindows()
```
